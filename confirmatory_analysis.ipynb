# Confirmatory Analysis

```python
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df = pd.read_csv("data.csv")

formulas = [
    "n_posts ~ week + is_from_rdepression * incidence",
    "n_posts_std ~ week_std + is_from_rdepression_std * incidence_std",
    "sentiment ~ week + is_from_rdepression * incidence",
    "sentiment_std ~ week_std + is_from_rdepression_std * incidence_std",
    "n_posts ~ week + is_from_rdepression * is_during_covid",
    "n_posts_std ~ week_std + is_from_rdepression_std * is_during_covid_std",
    "sentiment ~ week + is_from_rdepression * is_during_covid",
    "sentiment_std ~ week_std + is_from_rdepression_std * is_during_covid_std"
    ]

for formula in formulas:
    model = smf.ols(formula=formula, data=df).fit()
    print(f"Model formula: {formula}")
    print(model.summary())
```
